```python
import requests
import pandas as pd
import pprint

url = "https://intelifunctiongetdata.azurewebsites.net/api/InteliFunctionGetData"
params = {
    "code": "pZh3gmJW_87epswrWDuB7CvQle-KqjsVh2ZJUaifiXd4AzFuOEy98w==",
    "table": "sale",
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()

    # Criar um DataFrame com os dados
    df = pd.DataFrame(data)

    # Nome do arquivo CSV
    csv_filename = "dados_saida.csv"

    # Salvar DataFrame para um arquivo CSV
    df.to_csv(csv_filename, index=False)

    print(f"Dados salvos com sucesso no arquivo CSV: {csv_filename}")

    # Imprimir o conteúdo do arquivo CSV
    print("\nConteúdo do arquivo CSV:")
    df_from_csv = pd.read_csv(csv_filename)
    print(df_from_csv)

    # Pretty print dos dados
    pprint.pprint(data)
else:
    print("Falha ao obter os dados. Código de status:", response.status_code)
```
